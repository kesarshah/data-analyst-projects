# Employee Attrition Analysis
This notebook demonstrates data cleaning and exploratory analysis using Pandas and Matplotlib.

```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv('employee_dataset.csv')
df.head()
```

```
df['Attrition'].value_counts()
```

```
# Basic group analysis
by_dept = df.groupby('Department').agg(Total=('EmployeeID','count'), AttritionCount=('Attrition','sum'))
by_dept['AttritionRate'] = by_dept['AttritionCount'] / by_dept['Total']
by_dept
```

```
plt.figure(figsize=(6,4))
by_dept['AttritionRate'].plot(kind='bar')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.tight_layout()
```
